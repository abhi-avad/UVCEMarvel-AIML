```python
# Importing all dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
from sklearn import metrics
```

```python
# Accesing the dataset
iris = datasets.load_iris()
print(iris)
```

```python
iris.keys()
```

```python
print(iris['target_names'])
print(iris['target'])
print(iris['feature_names'])
print(iris['data'])

#print(iris.target_names)
#print(iris.target)
#print(iris.feature_names)
#print(iris.data)
```

```python
# Building the Dataframe
df = pd.DataFrame(iris.data)
df.columns = iris.feature_names
df['target'] = iris.target
#df_x = pd.DataFrame(iris.data)
#df_x.columns = iris.feature_names
#df_x.head()
```

```python
#df_y = pd.DataFrame(iris.target)
#df_y.head()
```

```python
#Splitting dataset int 70% training and 30% testing
x_train, x_test, y_train, y_test = train_test_split(df[iris.feature_names], df['target'], test_size=0.30, random_state=12)
print(x_train)
#print(x_test)
```

```python
# Logistic Regression Training
reg = LogisticRegression(max_iter=1000)
reg.fit(x_train,y_train)
```

```python
# Testing
y_pred = reg.predict(x_test)
```

```python
# Accuracy
reg.score(x_test, y_test)
```

```python
# DataFrame of Coefficients and Intercepts of Trained Model
reg_coefs = pd.DataFrame(reg.coef_, columns = ['sep_l_coef','sep_w_coef','pet_l_coef','pet_w_coef'])
reg_coefs['intercept'] = reg.intercept_
reg_coefs
```

```python
#Confusion Matrix
confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
#-------- IF YOU HAVE AN ISSUE SAYING DATA MUST BE 1D -> give y_test[0] as input instead of y_test
sns.heatmap(confusion_matrix, annot=True)
print('Accuracy: ',metrics.accuracy_score(y_test, y_pred))
plt.show()
```
